```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
```

```python
df_halda = pd.read_excel('halda.xlsx', na_values="x")
df_halda['target'] = 1
df_halda
```

```python
df_kontrola = pd.read_excel('kontrola.xlsx', na_values="x")
df_kontrola['target'] = 0
df_kontrola
```

```python
df_halda.shape, df_kontrola.shape
```

```python
df_halda.to_csv('proba_halda.csv', index=False)
```

```python
df_kontrola.to_csv('proba_porolny.csv', index=False)
```

```python
dataset = pd.concat([df_halda, df_kontrola])
```

```python
dataset
```

```python
dataset.fillna(0, inplace=True)
```

```python
dataset.to_csv('./prepared_datasets/based_dataset.csv', index=False)
```

```python
scaler = MinMaxScaler()

dataset_scaled = pd.DataFrame(scaler.fit_transform(dataset.values))
dataset_scaled.columns = dataset.columns
dataset_scaled
```

```python
dataset_scaled.to_csv('./prepared_datasets/normalized_dataset.csv', index=False)
```
